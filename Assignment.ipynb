```python
# Libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
from keras.datasets import fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
label_to_text = {0: 'T-shirt/top',
                 1: 'Trouser',
                 2: 'Pullover',
                 3: 'Dress',
                 4: 'Coat',
                 5: 'Sandal',
                 6: 'Shirt',
                 7: 'Sneaker',
                 8: 'Bag',
                 9: 'Ankle boot'}
```

```python
sample = []
for i in np.unique(y_train):
    result = np.where((y_train==i))[0]
    sample.append(x_train[result[0]])

np.sort(sample)
plt.figure(figsize = (15,7))

for i in range(10):
    plt.subplot(2,5,i+1)
    plt.imshow(sample[i], cmap=plt.get_cmap('gray'),label=[i])
    plt.title(label_to_text[i])
plt.suptitle("Fashion-MNIST Dataset")
plt.show()
```
